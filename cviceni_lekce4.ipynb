```python
import pandas
import numpy as np
uchazeci = pandas.read_csv("prijimaci_zkousky.csv")
uchazeci
```

Vytvoř funkci get_points(), která přidělí každému z uchazečů počet bodů na základě daných kritérií. 

```python
def get_points(row):
    points = 0
    if row[["body_aj", "body_mat", "body_cj"]].min() < 60:
        points = 0
    if row[["body_aj", "body_mat", "body_cj"]].min() >= 60:
        points = row[["body_aj", "body_mat", "body_cj"]].sum()
        if row["letni_skola"] == "ano":
            points += 5
        if pandas.notna(row["souteze"]):
            points += 10
    return points
uchazeci["score"] = uchazeci.apply(get_points, axis=1)
uchazeci.head(15)
```

Dále vytvoř novou tabulku, která obsahuje pouze uchazeče a uchazečky, kteří mají nenulové hodnocení. 

```python
uchazeci_filtered = uchazeci[uchazeci["score"] > 0].copy()
uchazeci_filtered
```

Dále využij metodu rank() k výpočtu pořadí uchazeče/uchazečky v rámci daného oboru, a data seřaď podle oboru a počtu bodů.

```python
uchazeci_filtered["rank"] = uchazeci_filtered.groupby("obor")["score"].rank(method="min", ascending=False)
uchazeci_filtered = uchazeci_filtered.sort_values(["obor", "score"], ascending=[True, False])
uchazeci_filtered.head(10)
```

Nakonec vytvoř tabulku s přijatými uchazeči, kde budou uchazeči hodnocení od 1. do 30. místa pro daný obor.

```python
uchazeci_filtered["prijat"] = np.where(uchazeci_filtered["rank"] <= 30, "prijat(a)", "neprijat(a)")
uchazeci_filtered
```
